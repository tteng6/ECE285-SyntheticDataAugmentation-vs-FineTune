### add suffix for each label

```python
from pathlib import Path
import shutil

INPUT_LABEL_DIR = Path(r'D:\UCSD_courses\ECE285\project\BG_preserved\labels')
suffixes = ['_frozen', '_itasha', '_rusty']

for label in INPUT_LABEL_DIR.iterdir():
    if label.suffix != '.txt':
        continue
    # create 3 copies with new names
    for s in suffixes:
        new_name = label.stem + s + '.txt'
        new_path = INPUT_LABEL_DIR / new_name
        shutil.copy2(label, new_path)
    # remove the original file
    label.unlink()
```
